```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books= pd.read_csv ('data/myDataset_1.csv')
```

```python
books.head ()
```

```python
 users= pd.read_csv ('data/userInfo.csv')
```

```python
users.head()
```

```python
finalBookMrg = users.merge(books, on= 'ISBN')
```

```python
finalBookMrg.head()
```

```python
books.head ()
```

```python
finalBookMrg.shape
```

```python
finalBookMrg.head()
```

```python
book_pivot = finalBookMrg.pivot_table(columns='user_id', index='title', values='ratings')
```

```python
book_pivot.head()
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[97,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_pivot.index
```

```python
books_name = book_pivot.index
```

```python
# saving the model and all necessary data()
import pickle
pickle.dump(model, open('OurData/model.pkl', 'wb'))
pickle.dump(books_name, open('OurData/books_name.pkl', 'wb'))
pickle.dump(finalBookMrg, open('OurData/finalBookMrg.pkl', 'wb'))
pickle.dump(book_pivot, open('OurData/book_pivot.pkl', 'wb'))
```

```python
# this is to check for the index exact thats why we add [0][0] 
np.where(book_pivot.index == "Introduction to Automata Theory")[0][0]
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i  in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
# below are two examples to verify that the model works
```

```python
# example_one

book_name = 'Introduction to Automata Theory'
recommend_book(book_name)
```

```python
# example_two

book_name = 'Software Engineering'
recommend_book(book_name)
```
